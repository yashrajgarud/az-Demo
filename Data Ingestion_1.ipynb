```
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("BRT").getOrCreate()

df = spark.read.format("csv") \
  .option("header", "true") \
  .option("sep", ",") \
  .load("dbfs:/FileStore/datafile")

display(df)
```

```
def castToInt(df,column):
        try:
               return df[column].cast(IntegerType())
        except ValueError:
              return 0
```

```
from pyspark.sql.types import *
df = df.withColumn("CastedAge",castToInt(df,"Age"))
```

```
df = df.withColumnRenamed('Age 	', 'Age')
```

```
from pyspark.sql.types import *
df.printSchema()
```

```
from pyspark.sql.functions import trim

df = df.withColumn("Name", trim(df.Name))
      .withColumn("Age", trim(df.Age)) \
      .withColumn("Date", trim(df.Date)) \
      .withColumn("Salary", trim(df.Salary))
```

```
from pyspark.sql import functions as f

for colname in df.columns:
    df = df.withColumn(colname, f.trim(f.col(colname)))
```
